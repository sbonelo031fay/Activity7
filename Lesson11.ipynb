# Set-up & Configuration


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#apply the default theme
sns.set_theme()

df = pd.read_csv("C:/Users/LENOVO/OneDrive/Documents/ITStudies/TP2/Activities/dataset - 2020-09-24.csv")

df.columns
df.dtypes
```

# Setting The Aesthetic Style

```python
# cross_accuracy = df['Cross accuracy %']
# jersey_number = df['Jersey Number']

#Setting different styles to compare
sns.set_style('whitegrid')

plt.figure(figsize=(10,7))
sns.histplot(x="Appearances", y="Assists", data=df)
plt.show()
```

```python
sns.set_style('ticks')

sns.lineplot(x='Wins', y='Losses', data=df)
plt.show()
```

```python
df.isnull().sum()
```

# Scaling Plots for Different Contexts

```python
# Setting the context to "talk" for larger plots suitable for presentations
plt.figure(figsize=(10, 7))
sns.set_context("talk")
sns.lineplot(x="Appearances", y="Assists", data=df,color='r')
plt.show()
```

```python
#Setting different styles to compare
sns.set_context()

plt.figure(figsize=(10,7))
sns.scatterplot(x="Appearances", y="Assists", data=df, color='r')
plt.show()
```
